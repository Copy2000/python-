# 1.创建ndarray：n direction array

## 1.使用np.array()由python list创建

```python
l = [1,2,3,4]
n = np.array(l)
display(l,n)
```

### numpy 默认ndarray的所有元素的类型是相同的
### 如果传进去的列表汇总包含不同的类型，则同一为同一类型，优先级：str>float>int

## 2.使用np中的routines函数创建

包括以下常见创建方法：
    1 np.ones(shape,dtype = None,order='C')
    2 np.zeros(shape,dtype = float,order = 'C')
    3 np.full(shape,fill_value,dtype = None,order = 'C')
    4 np.eye(N,M = None,k = 0,dtype=float)对角线为1其他位置全为0
    5 np.linspace(start,stop,num=50,endpoint=True,retstep=False,dtype=None)
    6 np.arange([start,]stop,[step,]dtype=None)
    7 np.random.randint(low,high=None,size=None,dtype='1')
    8 np.random.randn(d0,d1,d2,d3,...,dn)

```python
#全是1的ndarray
np.ones(shape=(3,4),dtype=np.float)
```

```python
#全是0的ndarray
np.zeros(shape=(2,3),dtype=np.int32)
```

```python
#使用指定的值去填充ndarray
np.full(shape=(5,5),fill_value=6,dtype=np.float16)
```

```python
#对角线为1其他位置全为0，k控制主对角线
np.eye(5,7,k=3)
```

```python
#endpoint表示最后一个是否显示,retstep显示间隔
np.linspace(0,100,num=100,endpoint=False,retstep=True)
```

```python
#跟python中发range一样
#左开右闭
#整数用arange，小数用linspace
np.arange(0,100,2)
```

```python
np.random.randint(0,150,size=(5,2,3))
```

```python
#正态分布(.mean是（）取平均值)
np.random.normal(loc=10,scale=3,size=(1000,3)).mean()
```

```python
#标准正态分布，方差为1，0-1
np.random.random(size=(10,10))
#np.random.rand(10,10)#和上面是一样的
```

```python
#-1 - 1
np.random.randn(2,3,4)
```
