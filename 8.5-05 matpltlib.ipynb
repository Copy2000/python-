## 灰色处理图片
因为在机器学习中色彩是没有用的

```python
jin = plt.imread('./金正恩.jfif')
plt.figure(figsize=(10,8))
plt.imshow(jin)
```

三种方法

```python
#最大值法
jin_max = jin.max(axis=2)
plt.figure(figsize=(10,8))
plt.imshow(jin_max,cmap='gray')
```

```python
#最小值
jin_min = jin.min(axis=2)
plt.figure(figsize=(10,8))
plt.imshow(jin_min,cmap='gray')
```

```python
#平均值
jin_mean = jin.mean(axis=2)
plt.figure(figsize=(10,8))
plt.imshow(jin_mean,cmap='gray')
```

```python
#加权平均:一般灰色处理取[0.299,0.587,0.114]
jin_weight = np.dot(jin,[0.299,0.587,0.114])/3
plt.figure(figsize=(10,8))
plt.imshow(jin_weight,cmap='gray')
```

```python
x,y=np.linspace(0,7,100),np.linspace(0,7,100)
X,Y= np.meshgrid(x,y)
```

```python
Z =np.sin(X)**2+np.cos(Y)**2+2
```

```python
plt.figure(figsize=(10,8))
threed = plt.subplot(projection='3d')
threed = threed.plot_surface(X,Y,Z,cmap='rainbow')
plt.colorbar(threed,shrink=0.8)
```
