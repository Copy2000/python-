## 直方图

一般用来表示数据的分布情况

```python
n = np.array([1,2,3,4,4,5,6,6,7,8])
s = pd.Series(data=n)
s.plot(kind='hist',bins=10,density=True)
#加入density=True之后，y轴的值变成了概率/组距的结果
#如果bins设置不太合理，可以加上kde图来显示数据分布
s.plot(kind='kde')#kernel density estimate
```

绘制一个由两个不同的正态分布组成的双峰分布

```python
n1 = np.random.normal(loc=5,scale=5,size=10000)
n2 = np.random.normal(loc=30,scale = 8,size=10000)

n = np.hstack((n1,n2))
s = pd.Series(data=n)
s.plot(kind = 'hist',bins=10,density=True)
s.plot(kind='kde')
```

## 散点图

散点图研究两个一维数据之间的关系的有效方法DataFrame对象可用

```python
df = pd.DataFrame({'A':np.random.randn(1000),'B':np.random.randn(1000),'C':np.random.randn(1000)})
df
```

```python
df.plot(x='A',y='B',kind='scatter')
```

```python
x= np.linspace(0,2*np.pi,100)
y = np.sin(x)
df2 = pd.DataFrame({'x':x,'y':y})
df2.plot(x='x',y='y')
```

散点图矩阵，当有多个点时，两两点的关系

使用函数：pd.plotting.scatter_matrix()

- 参数diagnol：设置对角线的图像类型

```python
pd.plotting.scatter_matrix(df,figsize=(16,16),alpha=0.6,diagonal='kde') #alpha是透明度,diagonal='kde'or 'hist'
```
