# 1.创建多层行索引
## 1） implicit building
常见方法是给dataframe构造函数的index参数传递两个或者更多的数组

```python
data = np.random.randint(0,150,size=(6,6))
index = [['class1','class1','class1','class2','class2','class2'],['张三','李四','王五','赵六','田七','孙八']]
columns = [['期中','期中','期中','期末','期末','期末'],['语文','数学','英语','语文','数学','英语']]
df = pd.DataFrame(data = data,index=index,columns=columns)
df
```

- Series创建多层索引

```python
index = [['class1','class1','class1','class2','class2','class2'],['张三','李四','王五','赵六','田七','孙八']]
data = np.random.randint(0,150,size=6)
s = pd.Series(data=data,index=index)
s
```

## 显示构造pd.MultiIndex
- 使用数组

```python
data = np.random.randint(0,150,size=(6,6))
index = pd.MultiIndex.from_arrays([['class1','class1','class1','class2','class2','class2'],['张三','李四','王五','赵六','田七','孙八']])
columns = pd.MultiIndex.from_arrays([['期中','期中','期中','期末','期末','期末'],['语文','数学','英语','语文','数学','英语']])
df = pd.DataFrame(data = data,index=index,columns=columns)
df
```

# 3.多层索引对象的索引与切片操作
1） Series的操作
【重要】 对于Series来说，直接中括号[]与使用loc()完全一样，推荐使用中括号索引和切片。

```python
s
```

- 索引

原则：当有多层索引的时候，不要直接使用索引内层索引

```python
#先索引外层索引
s.loc['class1','张三']
```

- 切片

```python
#不能直接切内层
s.loc['class1',:]
```

implicit 

```python
s.iloc[0]
```

```python
s.iloc[[0]]
```
