先将图片的类型从三维变成二维，reshape（60000，-1），-1的作用是将后面两个维度合并

```python
data = x_.reshape(60000,-1)[0:5000,:]
```

```python
data.shape
```

```python
target = y_[0:5000]
```

```python
target.shape
```

```python
x_test = x_1.reshape(10000,-1)
```

```python
y_test = y_1
```

引入knn

```python
knn = KNeighborsClassifier()
knn.fit(data,target)
```

```python
y_ = knn.predict(x_test)
```

```python
plt.figure(figsize=(10*2,10*3))
for i in range(100):
    axes = plt.subplot(10,10,i+1)
    axes.imshow(x_test[i*10].reshape(28,28),cmap='gray')
    axes.axis('off')
    if y_test[i*10]!=y_[i*10]:
        axes.set_title('True:%s\nPredict:%s' %(y_test[i*10],y_[i*10]),fontdict=dict(fontsize=15,c='r'))
    axes.set_title('True:%s\nPredict:%s' %(y_test[i*10],y_[i*10]))
        
```
